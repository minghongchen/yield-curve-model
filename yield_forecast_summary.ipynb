# U.S Treasury Yield Forecast

The project studies forecasts of U.S Treasury yields across different maturities based on different predictive models:

1. Vector autoregression (VAR) fitting principal components(PCA) of raw yields
2. VAR fitting Nelson-Siegel (NS) factors of raw yields
3. LSTM trained on raw yields


```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Out-of-sample forescast results

We use root mean square errors (RMSE) of forecasted results across all maturities and horizons as the measure of performance.

```python
horizons = [1,3,6,12]
var_pca_rmse = pd.read_csv('results/VAR-PCA_RMSE.csv')
var_ns_rmse = pd.read_csv('results/VAR-NS_RMSE.csv')
lstm_rmse = pd.read_csv('results/LSTM_RMSE.csv')

fig, ax = plt.subplots(2,2, figsize=(15,10))
for i in range(len(horizons)):
    r = int(i/2)
    c = i%2
    ax[r,c].plot(var_pca_rmse.iloc[i], label='VAR+PCA')
    ax[r,c].plot(var_ns_rmse.iloc[i], label='VAR+NS')
    ax[r,c].plot(lstm_rmse.iloc[i], label='LSTM')
    ax[r,c].legend()
    ax[r,c].set_xlabel('Maturities')
    ax[r,c].set_ylabel(f'RMSE ({horizons[i]}-month ahead)')
```
